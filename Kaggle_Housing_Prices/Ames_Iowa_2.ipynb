```python
## Create df with properties similar to the Blueste properties

combined_Blueste=data.loc[((data['BldgType']=='Twnhs')|(data['BldgType']=='TwnhsE'))&(data['CentralAir']=='Y')&(data['GrLivArea']>1200)&(data['GrLivArea']<1600)&(data['SalePrice']>110000)&(data['SalePrice']<165000)&(data['MSZoning']=='RM')]
combined_Blueste
```

```python
neigh_hoods= [CollgCr_df, Veenker_df, Crawfor_df, NoRidge_df, Mitchel_df, Somerst_df,
       NWAmes_df, OldTown_df, BrkSide_df, Sawyer_df, NridgHt_df, NAmes_df,
       SawyerW_df, IDOTRR_df, MeadowV_df, Edwards_df, Timber_df, Gilbert_df,
       StoneBr_df, ClearCr_df, NPkVill_df, Blmngtn_df, BrDale_df, SWISU_df,
       combined_Blueste]
```

```python
for each_df in neigh_hoods:
    name=each_df['Neighborhood'].iloc[0]
    binary_df=pd.get_dummies(each_df)
    
    y=binary_df[['SalePrice']].values.reshape(-1,1)

    x=binary_df.drop('SalePrice', axis=1)
    x=binary_df.fillna(0)
    
    X_train, X_test, y_train, y_test= train_test_split(x, y, random_state=23)
    
    model=LinearRegression()
    model.fit(X_train, y_train)
    train_score=model.score(X_train, y_train)
    test_score=model.score(X_test, y_test)
    
    print(f"{name}   Training Score: {train_score}    Test Score: {test_score}   Shape: {binary_df.shape}")
```
