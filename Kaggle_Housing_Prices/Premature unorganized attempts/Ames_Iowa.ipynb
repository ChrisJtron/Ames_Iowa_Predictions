## Prepare Data

```python
data=pd.read_csv("../Kaggle_Housing_Prices/train.csv")
data.head()
```

```python
binary_data=pd.get_dummies(data)
binary_data.head()
```

## Create X and Y data

```python
# create y value of SalePrice

y=binary_data[['SalePrice']].values.reshape(-1,1)
```

```python
# Create X values

xdf=binary_data.drop('SalePrice', axis=1)
xdf=xdf.drop('Id', axis=1)
xdf=xdf.fillna(0)
xdf
```

```python
X=xdf
print(X.shape)
print(y.shape)
```

##  Split data into training and testing


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=42)
```

## Create Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

# K Nearest Neighbor Model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# scale the model
X_scaler= StandardScaler().fit(X_train)
```

```python
Xtrain_scaled= X_scaler.transform(X_train)
Xtest_scaled= X_scaler.transform(X_test)
```

```python
#     knn=KNeighborsRegressor(n_neighbors=k, weights='distance')

for k in range(1, 15):
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(Xtrain_scaled, y_train)
    train_score=knn.score(Xtrain_scaled, y_train)
    test_score=knn.score(Xtest_scaled, y_test)
    print(f"K: {k},  TrainScore: {train_score}    TestScore: {test_score}")
```

```python
from sklearn.neighbors import NearestNeighbors
neigh=NearestNeighbors(n_neighbors=10)
neigh.fit(Xtest_scaled)

Xsample=xdf.iloc[32].values.reshape(1,-1)
print(neigh.kneighbors(Xsample))
print(knn.predict(Xsample))
```
